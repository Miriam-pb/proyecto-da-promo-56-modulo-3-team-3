FUNCIÓN CREAR BASE DE DATOS:

```python
def crear_base_datos(nombre_bd, user='root', password='AlumnaAdalab', host='127.0.0.1'):

    try:
        cnx = mysql.connector.connect(user=user, password=password, host=host)
        mycursor = cnx.cursor()
        sql = f"CREATE DATABASE {nombre_bd}"
        mycursor.execute(sql)
        print(f"✅ Base de datos `{nombre_bd}` creada correctamente.")
    except mysql.connector.Error as err:
        print(f"❌ Error al crear la base de datos `{nombre_bd}`.")
        print("Error Code:", err.errno)
        print("SQLSTATE:", err.sqlstate)
        print("Message:", err.msg)
    finally:
        mycursor.close()
        cnx.close()
```

```python
nombre_bd = "Better_Place_To_Work"
crear_base_datos(nombre_bd)
```

FUNCIÓN CREAR TABLAS:

```python
def crear_tablas(queries, user='root', password='AlumnaAdalab', host='127.0.0.1', database='Better_Place_To_Work'):

    try:
        cnx = mysql.connector.connect(user=user, password=password, host=host, database=database)
        mycursor = cnx.cursor()

        for sql in queries:
            mycursor.execute(sql)
            cnx.commit()
            print(f"✅ Tabla creada")

    except mysql.connector.Error as err:
        print(f"❌ Error al crear tablas.")
        print("Error Code:", err.errno)
        print("SQLSTATE:", err.sqlstate)
        print("Message:", err.msg)
    finally:
        mycursor.close()
        cnx.close()
```

```python
queries = [
        """
        CREATE TABLE IF NOT EXISTS empleados (
                employeenumber INT NOT NULL,
                datebirth INT NOT NULL,
                age INT NOT NULL,
                gender VARCHAR(6) NOT NULL,
                maritalstatus VARCHAR(8) NOT NULL,
                standardhours VARCHAR(10) NOT NULL,
                remotework VARCHAR(3) NOT NULL,
                overtime VARCHAR(3) NOT NULL,
                PRIMARY KEY (employeenumber)
                )
        """,
        """                
        CREATE TABLE IF NOT EXISTS laboral (
                employeenumber INT NOT NULL,
                department VARCHAR(30) NOT NULL,
                jobrole VARCHAR(20) NOT NULL,
                joblevel INT NOT NULL,
                jobinvolvement INT NOT NULL,
                jobsatisfaction INT NOT NULL,
                yearsincurrentrole INT NOT NULL,
                yearssincelastpromotion INT NOT NULL,
                yearswithcurrmanager INT NOT NULL,
                PRIMARY KEY (employeenumber),
                CONSTRAINT fk_empleados_laboral FOREIGN KEY (employeenumber) 
                        REFERENCES empleados(employeenumber) 
                        ON DELETE CASCADE 
                        ON UPDATE CASCADE
                )
        """,
        """                
        CREATE TABLE IF NOT EXISTS educación (
                employeenumber INT NOT NULL,
                education INT NOT NULL,
                educationfield VARCHAR(20) NOT NULL,
                PRIMARY KEY (employeenumber),
                CONSTRAINT fk_empleados_educacion FOREIGN KEY (employeenumber) 
                        REFERENCES empleados(employeenumber) 
                        ON DELETE CASCADE 
                        ON UPDATE CASCADE
                )
        """,
        """                
        CREATE TABLE IF NOT EXISTS desempeño (
                employeenumber INT NOT NULL,
                performancerating FLOAT NOT NULL,
                worklifebalance FLOAT NOT NULL,
                relationshipsatisfaction INT NOT NULL,
                environmentsatisfaction INT NOT NULL,
                trainingtimeslastyear INT NOT NULL,
                percentsalaryhike INT NOT NULL,
                educationfield VARCHAR(3) NOT NULL,
                PRIMARY KEY (employeenumber),
                CONSTRAINT fk_empleados_desempeño FOREIGN KEY (employeenumber) 
                        REFERENCES empleados(employeenumber) 
                        ON DELETE CASCADE 
                        ON UPDATE CASCADE
                )
        """,
        """                
        CREATE TABLE IF NOT EXISTS compensación (
                employeenumber INT NOT NULL,
                salary FLOAT NOT NULL,
                dailyrate FLOAT NOT NULL,
                stockoptionlevel INT NOT NULL,
                PRIMARY KEY (employeenumber),
                CONSTRAINT fk_empleados_compensacion FOREIGN KEY (employeenumber) 
                        REFERENCES empleados(employeenumber) 
                        ON DELETE CASCADE 
                        ON UPDATE CASCADE
                )
        """,
        """                
        CREATE TABLE IF NOT EXISTS experiencia (
                employeenumber INT NOT NULL,
                totalworkingyears FLOAT NOT NULL,
                numcompaniesworked INT NOT NULL,
                yearsatcompany INT NOT NULL,
                distancefromhome INT NOT NULL,
                businesstravel VARCHAR(15) NOT NULL,
                PRIMARY KEY (employeenumber),
                CONSTRAINT fk_empleados_experiencia FOREIGN KEY (employeenumber) 
                        REFERENCES empleados(employeenumber) 
                        ON DELETE CASCADE 
                        ON UPDATE CASCADE
                )
        """
    ]
crear_tablas(queries)
```

FUNCIÓN INSERTAR DATOS:

```python
def insertar_datos(sql, valores, user='root', password='AlumnaAdalab', host='127.0.0.1', database='Better_Place_To_Work'):

    try:
        cnx = mysql.connector.connect(user=user, password=password, host=host, database=database)
        mycursor = cnx.cursor()
        mycursor.executemany(sql, valores)
        cnx.commit()
        print(f"✅ {mycursor.rowcount} registro(s) insertado(s).")
    except mysql.connector.Error as err:
        print(f"❌ Error al insertar datos.")
        print("Error Code:", err.errno)
        print("SQLSTATE:", err.sqlstate)
        print("Message:", err.msg)
    finally:
        mycursor.close()
        cnx.close()
```

```python
sql = "INSERT INTO empleados (employeenumber, datebirth, age, gender, maritalstatus, standardhours, remotework, overtime) VALUES (%s, %s, %s, %s, %s, %s, %s, %s)"
valores = valores_empleados
insertar_datos(sql, valores)
```
