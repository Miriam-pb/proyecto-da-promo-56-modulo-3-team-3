```python
import pandas as pd
import numpy as np
pd.set_option("display.max_columns", None)
```

```python
columnas1_10 = pd.read_csv("columnas1-10.csv", index_col=0)
columnas11_20 = pd.read_csv("columnas11-20.csv")
columnas21_30 = pd.read_csv("columnas21-30.csv")
columnas31_40 = pd.read_csv("columnas31-40.csv", index_col=0)
df_numcompaniesworked = pd.read_csv('hr_raw_data.csv', index_col=0) # para añadir la columna extraviada
```

```python
# extraigo la columna:
columna_numcompaniesworked = df_numcompaniesworked[['numcompaniesworked']]
columna_numcompaniesworked
```

```python
csv_completo = pd.concat(
    [columnas1_10, columnas11_20, columna_numcompaniesworked, columnas21_30, columnas31_40],
    axis=1,
    ignore_index=False
)
```

```python
csv_completo.columns = csv_completo.columns.str.lower() # columnas en minusculas
```

```python
# Detectar columnas de tipo texto
columnas_texto = csv_completo.select_dtypes(include='object').columns

# Aplicar .str.lower() solo a esas columnas
csv_completo[columnas_texto] = csv_completo[columnas_texto].apply(lambda col: col.str.lower())
```

```python
csv_completo.head()
```

```python
csv_completo.to_csv("csv_completo.csv", index=False)
```
