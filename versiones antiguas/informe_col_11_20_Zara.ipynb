## 2.1. Limpieza variable maritalstatus

```python
cols_zara["maritalstatus"].unique() # Vemos los valores únicos de la var.
```

```python
# Primero, cambiamos todos los valores a minúsculas:
cols_zara["maritalstatus"] = cols_zara["maritalstatus"].str.lower()
```

```python
# Segundo, corregimos los valores en los que hemos observado errores: 
cols_zara["maritalstatus"]=cols_zara["maritalstatus"].str.replace("marreid", "married")
```

```python
cols_zara["maritalstatus"].unique() # Comprobamos que ya no existe "marreid" como valor único. 
```

```python
cols_zara["maritalstatus"] = cols_zara["maritalstatus"].str.strip() # Espacios en blanco al inicio y final de cada valor en la columna "maritalstatus".
```

```python
cols_zara["maritalstatus"]
```

## 2.4. Limpieza variable environmentsatisfaction

```python
cols_zara["environmentsatisfaction"] # Vemos que los valores son de 1 a 49. Confirmamos con César que se trata de un error y la escala debe ser de 1 a 4.
```

```python
# Primero pasamos la variable a string
cols_zara["environmentsatisfaction"] = cols_zara["environmentsatisfaction"].astype(str)
```

```python
cols_zara["environmentsatisfaction"].dtype # Comprobamos que la transformación se ha realizado correctamente.
```

```python
# Luego, usamos regex para cambiar cualquier valor que empieza por el número de la escala a ese mismo número. Así, los valores que comienzan por 1 (12, 16, etc.) se transforman a 1, y así de manera sucesiva con los demás números.
cols_zara["environmentsatisfaction"] = cols_zara["environmentsatisfaction"].replace({r"^1": 1, r"^2": 2, r"^3": 3, r"^4": 4}, regex=True)
```

```python
# Hacemos una comprobación de los valores únicos:
cols_zara["environmentsatisfaction"].unique()
```

```python
# Volvemos a pasar los valores a int:
cols_zara["environmentsatisfaction"] = cols_zara["environmentsatisfaction"].astype(int)
```

```python
# Comprobamos el tipo de la var para asegurarnos de que sea int de nuevo.
cols_zara["environmentsatisfaction"].dtype
```

## 2.5. Limpieza variable hourlyrate

```python
# Vemos que podemos sacar los valores de hourlyrate a partir de la variable dailyrate de df. Los valores disponibles nos indican que hourlyrate=dailyrate/8
```

```python
cols_zara['dailyrate'] = df['dailyrate'] # Creamos una nueva columna en cols_zara a partir de la col dailyrate de df.
```

```python
cols_zara[["dailyrate", "hourlyrate"]] # Comparamos las dos columnas para asegurarnos de que el patrón es el que detallábamos arriba (hourlyrate=dailyrate/8). 
```

```python
cols_zara["dailyrate"].dtype # Comprobamos el tipo de var.
```

```python
cols_zara["hourlyrate"].dtype # Comprobamos el tipo de var
```

```python
cols_zara['hourlyrate'] = cols_zara['hourlyrate'].fillna(cols_zara['dailyrate'] / 8) # modificamos los nulos de la col hourlyrate a partir de los valores correspondientes de dailyrate dividido entre 8.
```

```python
cols_zara[["hourlyrate", "dailyrate"]] # comparamos las dos columnas y vemos que los valores null de hourlyrate ahora contienen los datos según lo establecido arriba.
```

```python
# Redondeamos a 2 decimales:

cols_zara["hourlyrate"]=round(cols_zara["hourlyrate"], 2)
cols_zara["dailyrate"]=round(cols_zara["dailyrate"], 2)
```

```python
# cols_zara = cols_zara.drop(columns='dailyrate') # borramos la columna dailyrate
# no se hace dado que usaremos la var en la limpieza de la siguiente var.
```

## 2.6. Limpieza variable monthlyrate

```python
# No tiene valores nulos, vemos que concuerda con los valores de dailyrate, siguiendo el siguiente patrón: monthlyrate=dailyrate*21.
# No obstante, requiere limpieza (cambiar tipo de valores a float, quitar símbolo de dólar, y convertir comas en puntos).
```

```python
cols_zara["monthlyrate"].sample(10)
```

```python
cols_zara["monthlyrate"].dtype # Comprobamos el tipo de var
```

```python
# Pasamos monthlyrate de object a float. Para ello, necesitamos quitar el símbulo del $ y cambiar las comas de los decimales a puntos.

cols_zara["monthlyrate"] = cols_zara["monthlyrate"].str.replace('$', '', regex=False)
cols_zara["monthlyrate"] = cols_zara["monthlyrate"].str.replace(',', '.', regex=False)
cols_zara['monthlyrate'] = cols_zara['monthlyrate'].astype(float)
```

```python
# Aunque en un principio no parece necesario, redondeamos a 2 decimales:

cols_zara["monthlyrate"]=round(cols_zara["monthlyrate"], 2)
```

```python
cols_zara = cols_zara.drop(columns='dailyrate') # Tras hacer las comprobaciones necesarias, borramos la columna dailyrate, dado que ya no la necesitamos en cols_zara.
```

## 2.7. Limpieza variable monthlyincome

```python
# Tras explorar df, hemos visto que monthlyincome (y también sameasmonthlyincome)= salary / 12. Salary no está en cols_zara, así que es necesario copiarla de df.
```

```python
cols_zara['salary'] = df['salary'] # Creamos la nueva variable en cols_zara a partir de la variable salary de df.
```

```python
cols_zara[['monthlyincome', "salary"]] # Confirmamos la relación de las variables de acuerdo con lo detallado arriba.
```

```python
cols_zara['monthlyincome'].dtype # Vemos que la variable es de tipo objeto.
```

```python
cols_zara['salary'].dtype # Vemos que la variable es de tipo objeto.
```

```python
# Pasamos los valores de monthlyrate de object a float. Para ello, primero necesitamos quitar el símbilo del $ y cambiar las comas de los decimales a puntos.
cols_zara["monthlyincome"] = cols_zara["monthlyincome"].str.replace('$', '', regex=False)
cols_zara["monthlyincome"] = cols_zara["monthlyincome"].str.replace(',', '.', regex=False)
cols_zara['monthlyincome'] = cols_zara['monthlyincome'].astype(float)
cols_zara['monthlyincome']
```

```python
# Pasamos los valores de salary de object a float. Para ello, primero necesitamos quitar el símbilo del $ y cambiar las comas de los decimales a puntos.
cols_zara["salary"] = cols_zara["salary"].str.replace('$', '', regex=False)
cols_zara["salary"] = cols_zara["salary"].str.replace(',', '.', regex=False)
cols_zara['salary'] = cols_zara['salary'].astype(float)
cols_zara['salary']
```

```python
# Rellenamos los valores nulos de la columna 'monthlyincome' calculando el ingreso mensual como 'salary / 12'.
cols_zara['monthlyincome'] = np.where(cols_zara['monthlyincome'].isna(), cols_zara['salary'] / 12, cols_zara['monthlyincome'])
```

```python
cols_zara[['monthlyincome', "salary"]]
```

```python
# Redondeamos a 2 decimales:

cols_zara["monthlyincome"]=round(cols_zara["monthlyincome"], 2)
cols_zara["salary"]=round(cols_zara["salary"], 2)
```

```python
cols_zara = cols_zara.drop(columns='salary') # Borramos la col "salary", dado que ya no la necesitamos.
```
